```python
import pandas as pd
import seaborn as sbs
import matplotlib.pyplot as plot
```

```python
%matplotlib inline
pkmn=pd.read_csv('./Pokemon.csv')
```

```python
sbs.jointplot(x='HP',y='Attack',data=pkmn,size=10,kind='reg')
```

```python
sbs.boxplot(data=pkmn)
pkmn=pd.melt(pkmn,id_vars=["Name","Type 1","Type 2"],var_name="Stats")
```

```python


sbs.swarmplot(x="Stats", y="value", data=pkmn, hue="Type 1");

```
