```python
pkmn=pd.read_csv('./Pokemon.csv');
columns=pkmn.columns.values
columns[0]='id'
pkmn.columns=columns
pkmn.head()
def generation(x):
    if x<152:
        return 1
    elif x>151 and x<252:
        return 2
    elif x>251 and x<387:
        return 3
    elif x>386 and x<494:
        return 4
    elif x>493 and x<650:
        return 5
    elif x>649 and x<722:
        return 6
    else:
        return 0
pkmn['gen']=map(generation,pkmn['id'])

pkmn.gen.value_counts(normalize=True) #normalized counting range from 0 to 1
pkmn.describe()
pkmn_mean=(pkmn.describe().iloc[1,2:8])
pkmn_mean.plot(kind="bar",width = 0.5,color=['red','orange','yellow','blue','green','pink'])
```
